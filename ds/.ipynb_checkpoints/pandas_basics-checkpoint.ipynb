```python
# Importing necessary libraries
import numpy as np
import pandas as pd

print("imported numpy and pandas")
```

```python
# Create a Series
data = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
print(data)
```

```python
# Create a DataFrame
'''
 you can also extract data from a csv or excel file
using pd.read_csv("file.csv") and pd.read_excel("file.xlsx") respectively.
'''
data = {
    "Name": ["Alice", "Bob", "Charlie","Abhinav"],
    "Age": [25, 30, 35,19],
    "City": ["New York", "Los Angeles", "Chicago","Mumbai"]
}
df = pd.DataFrame(data)
print(df)
```

```python
print("\nfirst 2 lines")
print(df.head(2))
print("\nlast 2 lines")
print(df.tail(2))
print("\nheaders")
print(df.columns)
```

```python
### Access a column
print(df["Name"])

print("\n\n")
# Access a row by index
print(df.iloc[-1])  # last row

print("\n\n")
# Access a row by label
print(df.loc[0])  # First row

print("\n\n")
# Access a specific (R,C)
print(df.iloc[0, 2])  # First row, third column
```

```python
# Filter rows where Age > 25
filtered_df = df.loc[df["Age"] > 25]
print(filtered_df)
```

```python
# Add a new column
df["Salary"] = [50000, 60000, 70000,80000]
print(df)
```

```python
print("\nDescription\n")
print(df.describe())  # Summary statistics
print("\nInfo\n")
print(df.info())      # DataFrame information
```

```python
# Sort by Age
df_sorted = df.sort_values(by="Age", ascending=True)
print(df_sorted)
```

```python
# Group by a column and calculate the mean

df2=pd.DataFrame({"Name":["p1","p2"], "Age":[0,0], "City":["Mumbai","Mumbai"],"Salary":[50,50]})

df=pd.concat([df,df2])

print("dropping duplicates cause concatinating adds every time u run")
df = df.drop_duplicates()
print(df)

print("\n")

grouped = df.groupby("City")["Age"].mean() #basically gives mean age of rows with a particular city
print(grouped)
```

```python
# conditional changes
df.loc[df["Age"] < 20, "Salary"] = 100000
print(df)
```

```python
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"A": [5, 6], "B": [7, 8]})
result = pd.concat([df1, df2])
print(result)
```

```python
df1 = pd.DataFrame({"ID": [1, 2], "Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"ID": [1, 2], "Age": [25, 30]})
merged = pd.merge(df1, df2, on="ID")
print(merged)
```
